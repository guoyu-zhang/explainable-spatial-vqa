```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

def draw_boxes(image_path, bboxes):
    """
    Draw bounding boxes on an image in a Jupyter notebook.

    Parameters:
    - image_path: str, path to the PNG image.
    - bboxes: list of bounding box coordinates in normalized format
              [x_min, y_min, x_max, y_max] (values between 0 and 1).
    """
    # Open the image using PIL
    img = Image.open(image_path)
    img_width, img_height = img.size  # Get image dimensions

    # Create a Matplotlib figure
    fig, ax = plt.subplots(1, figsize=(10, 7))

    # Display the image
    ax.imshow(img)

    # Draw each bounding box
    for bbox in bboxes:
        # Convert normalized coordinates to pixel coordinates
        x_min = bbox[0] * img_width
        y_min = bbox[1] * img_height
        x_max = bbox[2] * img_width
        y_max = bbox[3] * img_height

        # Compute box width and height
        width = x_max - x_min
        height = y_max - y_min

        # Add a rectangle to the plot
        rect = patches.Rectangle(
            (x_min, y_min), width, height, linewidth=2, edgecolor="r", facecolor="none"
        )
        ax.add_patch(rect)

    # Add labels, grid, and axis options
    ax.set_title("Bounding Boxes")
    ax.set_xlabel("X (pixels)")
    ax.set_ylabel("Y (pixels)")
    plt.grid(True)

    # Show the plot in the Jupyter notebook
    plt.show()


# Example usage in a Jupyter notebook:
# Replace "example_image.png" with the path to your image file.
image_path = "/Users/guoyuzhang/University/Y5/diss/vqa/code/data/CLEVR_v1.0/images/train/CLEVR_train_000001.png"  # Update with your PNG file

# Example bounding boxes in normalized coordinates
bboxes = [[0.45249182, 0.15373772, 0.5558415,  0.30876228],
          [0.39049593, 0.5419939, 0.47617075, 0.6705061 ],
          [0.6403395, 0.4198842, 0.7513272, 0.5863658 ],
          [0.5619498, 0.53445995, 0.6922169, 0.7718076 ],
          [0.5144942, 0.27799124, 0.5771725, 0.37200877],
          [0.77793473, 0.30127704, 0.93039864, 0.529973  ],
          [0.7587688, 0.4881532,  0.8412312, 0.6118468 ],
          [0.31046227, 0.30944338, 0.37703773, 0.40930662],
          [0.2332673, 0.36552596, 0.32506603, 0.503224 ]]

# Draw the bounding boxes
draw_boxes(image_path, bboxes)
```
